```python
#getting a dictionary from a text file
#Getting a dictionary from a CSV file
import  csv
with open('treeorderssubsetnodupes.csv', mode='r') as infile:
    reader = csv.reader(infile)
   #creating a dictionary - walk through code.
   # And introduce the idea of a Case Study
    mydict ={}

    for row in reader:
        key = row[0]
        mydict[key]=row[1]
```

```python
#Get the size of the dictonary
sizeOfDict = len(mydict)
print (sizeOfDict)
```

```python
#Give an option for the troop change the number of trees collected
mydict['205'] = 10
print(mydict)
```

```python
#Give an option for the troop to add a new customer
mydict['999'] = 12
print (mydict)
```

```python
#Print the contents of the dictionary
for i in mydict:
    print (i, mydict[i])
```
